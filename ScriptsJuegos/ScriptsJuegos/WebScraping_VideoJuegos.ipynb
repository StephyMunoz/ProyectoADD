```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import sqlite3
```

```python
def find_2nd(string, substring):
    return string.find(substring, string.find(substring) + 1)
def find_1st(string, substring):
    return string.find(substring, string.find(substring))   

response = requests.get("https://newzoo.com/insights/rankings/top-10-countries-by-game-revenues")
soup = BeautifulSoup(response.content, "lxml")

Data=[]
Ranking=[]
```

```python
post_ranking=soup.find_all("div", class_="rank-number")
post_table=soup.find_all("td", class_="")
```

```python
for element in post_table:
    element=str(element)
    limpio=str(element[find_1st(element, '>')+1:find_2nd(element, '<')])
    Data.append(limpio.strip())

for element in post_ranking:
    element=str(element)
    limpio=str(element[find_1st(element, '>')+1:find_2nd(element, '<')])
    Ranking.append(limpio.strip())
```

```python
dataJson={}
index=1
for i in range(100):
    try:
        dataJson[i] = {
            'Ranking': Ranking[i],
            'Country': Data[index],
            'Region': Data[index+1],
            'Population': Data[index+2],
            'IntPop': Data[index+3],
            'Revenues': Data[index+4]
        }
        index+=6
        
    except Exception as e:
        print("error: " + str(e))
```

```python
dfData = pd.DataFrame(dataJson)
```

```python
con=sqlite3.connect("videojuegos.db")
dfData.to_sql("videojuegos", con)
```
